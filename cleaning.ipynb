```python
from math import isnan

import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("Customer Call List.xlsx") #reading and storing the excel
df
```

```python
df = df.drop_duplicates() #dropping all instance of dups(exact same)
```

```python
df = df.drop(columns="Not_Useful_Column") #dropping the col
```

```python
df["Last_Name"] = df["Last_Name"].str.strip("._/") #stripping all instance of . _ and /
df
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace(r"[^a-zA-Z0-9]", "", regex=True) #getting rid of anything thats not alphanum
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x)) #converting all row into str
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:]) #mutating
df["Phone_Number"] = df["Phone_Number"].replace("nan--", np.nan)
df["Phone_Number"] = df["Phone_Number"].replace("Na--", np.nan)
df
```

```python
df[["Street_Address", "State", "Zip"]] = df["Address"].str.split(',', n=2, expand=True)
df.drop(columns=["Address"], inplace=True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace("Yes", "Y")
df["Paying Customer"] = df["Paying Customer"].str.replace("No", "N")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("Yes", "Y")
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace("No", "N")
```

```python
df = df.replace("NaN", np.nan) #Does nothing as "NaN" is not a value
df = df.replace("N/a", np.nan)
```

```python
df.fillna(np.nan, inplace=True) #filling all empty with NaN
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("N") #Filling all NaN with "N"
```

```python
for i in df.index:
    if df.loc[i, "Do_Not_Contact"] == "Y": #dropping all rows w/ "Y"
        df.drop(i, inplace=True)
    elif pd.isna(df.loc[i, "Phone_Number"]):
        df.drop(i, inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True) #Need to add drop=True, or it will save old index
df
```

```python
df.to_csv("cleaned_list.csv", index=False)
# df.to_excel("cleaned_list.xlsx", index=False)
```
