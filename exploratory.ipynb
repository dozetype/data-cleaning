```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("world_population.csv")
pd.set_option('display.float_format', '{:.2f}'.format)
df
```

```python
print(df.describe(), end='\n\n\n\n\n') #gives basic statistics of the data
print(df.info()) #gives overview of the data
```

```python
#Number of NaN for each Column
df.isnull().sum() #isnull() checks if the row is null, sum() adds up the rows that are true
```

```python
#Number of distinct value for each Column
df.nunique()
```

```python
#Sort the df base on 1 Column
# df.sort_values(by=['2020 Population'], ascending=False).head(10)

#Correlation between Columns (Numeric Value Only)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu')
plt.rcParams['figure.figsize'] = (17, 8)
plt.show()
```

```python
df.groupby("Continent").mean(numeric_only=True)
```

```python
df[df["Continent"].str.contains("Oceania")].sort_values(by=["2020 Population"], ascending=False)
```

Show Population in each continent over time

```python
df2 = df.groupby("Continent")[df.columns[5:13]].mean() #Get mean of relevant Cols
df2 = df2[df2.columns[::-1]] #Reverse the years
df2 = df2.transpose() #Make the years the rows, Continent the Col
df2.plot()
```
